```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
plt.style.use('default')
```

Para cada métrica: 

    - Diagramas de barras agrupadas, en horizontal.
    - Cada diagrama un dataset
    - Cada barra un método

```python
def plot_metric(methods, metric_name):
    index=["PlantVillage", "UTKFaceBias", "UTKFaceFull"]
    df = pd.DataFrame({'Baseline': methods['Baseline'],
                        'WGAN': methods['WGAN'],
                        'Loss Focal': methods['Loss Focal'],
                        'Loss Focal Difusa': methods['Loss Focal Difusa'],
                        'PRM-IM': methods['PRM-IM'],
                        'CoSen CNN': methods['CoSen CNN'],
                      }, index=index)
    ax = df.plot.bar(rot=0, color={
        'Baseline': "white",
        'WGAN': "green",
        'Loss Focal': "red",
        'Loss Focal Difusa': "white",
        'PRM-IM': "green",
        'CoSen CNN': "red",
    },edgecolor="black", figsize=(10,5))
    DASH = "/////"
    for bar, hatch in zip(ax.patches, [None,None,None,DASH,DASH,DASH]*3):
        bar.set_hatch(hatch)
    plt.legend(ncol=3)
    plt.ylim(0.5, 1.2)
    plt.ylabel(metric_name)
    plt.show()
```

# PlantVillage-UTKFaceBias-UTKFaceFull
